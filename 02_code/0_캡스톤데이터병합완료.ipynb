```python
comm = pd.read_csv('서울시_우리마을가게_상권분석서비스(구_상권_직장인구).csv', usecols=[1,4,5,6,10,11,12], encoding='cp949')
comm = pd.merge(mapp[['행정동_코드', '상권_코드']], comm, how='inner', on='상권_코드')
comm = comm.groupby(by=['행정동_코드']).mean()
comm.reset_index(inplace=True)
comm.drop(columns=['상권_코드', '기준_분기_코드'], inplace=True)
comm
```

데이터 병합 (dong, inhelp, budget, peo, gyn, comm)

```python
df = pd.merge(dong, gyn[['행정동', '공동육아나눔터유무']], how='outer', left_on='행정동명', right_on='행정동').drop('행정동', axis=1).fillna(int(0))
df = pd.merge(df, inhelp, left_on='시군구명', right_on='자치구').drop('자치구', axis=1)
df = pd.merge(df, budget, left_on='시군구명', right_on='자치구').drop('자치구', axis=1)
df = pd.merge(df, peo, left_on='행자부행정동코드', right_on='행정동코드').drop('행정동코드', axis=1)
df = pd.merge(df, comm, how='outer', left_on='행자부행정동코드', right_on='행정동_코드').drop('행정동_코드', axis=1)
df
```
